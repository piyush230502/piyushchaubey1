```python
#univariate analysis
# what is the distribution of age among heart failure patients in the datasets

plt.figure(figsize = (10, 6))
sns.histplot(x =df["DEATH_EVENT"],y = df["age"] , kde = True, color = 'green')
plt.title('Distribution of age')
plt.xlabel('death_event')
plt.ylabel('age')
plt.show()
```
